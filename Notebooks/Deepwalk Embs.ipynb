```python
fin_data = {}
```

```python
fout = "graph.adjlist"
fo = open(fout, "w")

for k, v in fin_data.items():
    fo.write(str(k) + ' ')
    for it in v:
        fo.write(str(it) + ' ')
    fo.write('\n')

fo.close()
```

```python
with open('graph.embeddings','r') as f:
    data = f.readlines()
```

```python
data = data[1:]
```

```python
fin_lines = []
for item in data:
    fin_lines.append(item[:-1])
```

```python
fin_lines
```

```python
deepwalk_embs = {}
```

```python
for rec in fin_lines:
    cur = rec.split(' ')
    
    cur_list = list(map(float, cur[1:]))
    deepwalk_embs[cur[0]] = cur_list
#     print(cur_list)
#     print(cur)
```
